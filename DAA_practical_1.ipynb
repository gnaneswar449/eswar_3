```python
import time

def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        swapped = False
        # Last i elements are already in place, so we don't check them
        for j in range(0, n - i - 1):
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True

        # If no two elements were swapped by inner loop, the array is sorted
        if not swapped:
            break
    return arr

def selection_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Find the minimum element in remaining unsorted array
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j

        # Swap the found minimum element with the first element of the unsorted part
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

if __name__ == "__main__":
    # 1. Take user input
    print("--- Sorting Algorithm Tester ---")
    user_input = input("Enter a list of numbers separated by spaces: ")

    # 2. Convert the input string into a list of integers
    try:
        original_array = [int(x) for x in user_input.split()]
    except ValueError:
        print("Invalid input! Please enter only integers separated by spaces.")
        exit()

    print(f"\nOriginal Array: {original_array}")
    print("-" * 30)

    # 3. Execute and measure Bubble Sort
    # We copy the array so both algorithms sort the exact same unsorted data
    arr_for_bubble = original_array.copy()

    start_time_bubble = time.perf_counter()
    bubble_sorted = bubble_sort(arr_for_bubble)
    end_time_bubble = time.perf_counter()

    bubble_exec_time = end_time_bubble - start_time_bubble

    print(f"Bubble Sort Result: {bubble_sorted}")
    print(f"Bubble Sort Execution Time: {bubble_exec_time:.8f} seconds")
    print("-" * 30)

    # 4. Execute and measure Selection Sort
    arr_for_selection = original_array.copy()

    start_time_selection = time.perf_counter()
    selection_sorted = selection_sort(arr_for_selection)
    end_time_selection = time.perf_counter()

    selection_exec_time = end_time_selection - start_time_selection

    print(f"Selection Sort Result: {selection_sorted}")
    print(f"Selection Sort Execution Time: {selection_exec_time:.8f} seconds")
    print("-" * 30)
```

```python
import time

def merge_sort(arr):
    # Base case: A list of zero or one elements is already sorted
    if len(arr) <= 1:
        return arr

    # 1. Divide: Find the middle point and split the array
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # 2. Conquer: Recursively sort both halves
    sorted_left = merge_sort(left_half)
    sorted_right = merge_sort(right_half)

    # 3. Combine: Merge the sorted halves together
    return merge(sorted_left, sorted_right)

def merge(left, right):
    sorted_arr = []
    i = j = 0

    # Compare elements from both sides and add the smaller one to the final list
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_arr.append(left[i])
            i += 1
        else:
            sorted_arr.append(right[j])
            j += 1

    # If there are any remaining elements in left or right, append them
    sorted_arr.extend(left[i:])
    sorted_arr.extend(right[j:])

    return sorted_arr

if __name__ == "__main__":
    # Take user input
    print("--- Merge Sort Algorithm Tester ---")
    user_input = input("Enter a list of numbers separated by spaces: ")

    # Convert input string into a list of integers
    try:
        original_array = [int(x) for x in user_input.split()]
    except ValueError:
        print("Invalid input! Please enter only integers separated by spaces.")
        exit()

    print(f"\nOriginal Array: {original_array}")
    print("-" * 35)

    # Execute and measure Merge Sort
    start_time = time.perf_counter()
    merge_sorted_result = merge_sort(original_array)
    end_time = time.perf_counter()

    exec_time = end_time - start_time

    print(f"Merge Sort Result: {merge_sorted_result}")
    print(f"Merge Sort Execution Time: {exec_time:.8f} seconds")
    print("-" * 35)
```

```python
import time

def quick_sort(arr):
    # Base case: arrays with 0 or 1 elements are already sorted
    if len(arr) <= 1:
        return arr
    else:
        # Using the last element as the pivot
        pivot = arr[-1]

        # Partitioning the array
        left_sub_array = [x for x in arr[:-1] if x <= pivot]
        right_sub_array = [x for x in arr[:-1] if x > pivot]

        # Recursively sorting the sub-arrays and combining them with the pivot
        return quick_sort(left_sub_array) + [pivot] + quick_sort(right_sub_array)

if __name__ == "__main__":
    # Take user input
    print("--- Quick Sort Algorithm Tester ---")
    user_input = input("Enter a list of numbers separated by spaces: ")

    # Convert input string into a list of integers
    try:
        original_array = [int(x) for x in user_input.split()]
    except ValueError:
        print("Invalid input! Please enter only integers separated by spaces.")
        exit()

    print(f"\nOriginal Array: {original_array}")
    print("-" * 35)

    # Execute and measure Quick Sort
    start_time = time.perf_counter()
    quick_sorted_result = quick_sort(original_array)
    end_time = time.perf_counter()

    exec_time = end_time - start_time

    print(f"Quick Sort Result: {quick_sorted_result}")
    print(f"Quick Sort Execution Time: {exec_time:.8f} seconds")
    print("-" * 35)
```
